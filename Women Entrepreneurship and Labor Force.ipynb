```python
base_numerical = base[['Women Entrepreneurship Index', 'Entrepreneurship Index',
       'Inflation rate', 'Female Labor Force Participation Rate']]
```

```python
base_categorical = base[['Country', 'Level of development', 'European Union Membership',
       'Currency']]
```

```python
base_numerical.head()
```

```python
base_categorical.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
hot = OneHotEncoder()

base_categorical = hot.fit_transform(base_categorical).toarray()
base_categorical = pd.DataFrame(base_categorical)
```

```python
base_categorical.head()
```

```python
base_all = pd.concat([base_categorical, base_numerical], axis = 1)
```

```python
X = base_all
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 4, init='k-means++', random_state=0)
y_kmeans = kmeansmodel.fit_predict(X)
```

```python
y_kmeans
```

```python
y_kmeans.shape
```

```python
base['cluster'] = y_kmeans
```

```python
base.head(30)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

principal_comp = pca.fit_transform(X)
```

```python
principal_comp
```

```python
pca_df = pd.DataFrame(data = principal_comp, columns=['pca1', 'pca2'])
pca_df.head()
```

```python
pca_df = pd.concat([pca_df, pd.DataFrame({'cluster': y_kmeans})], axis = 1)
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'pca1', y = 'pca2', hue = 'cluster', data = pca_df, palette = ['red', 'green', 'blue', 'pink'])
```
