```python
import pandas as pd # Manipulation Data
import numpy as np # Manipulation Arrays
from sklearn.tree import DecisionTreeRegressor #Decision Tree Model
from sklearn.tree import plot_tree # 
from sklearn.datasets import load_diabetes # Dataset
from sklearn.metrics import mean_squared_error # Loss
from sklearn.model_selection import train_test_split # Train and Test Dataset
%pylab inline
```

```python
X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
#tree = DecisionTreeRegressor(max_depth=1)
tree = DecisionTreeRegressor(min_samples_leaf=10, random_state=0)

tree.fit(X_train, y_train)
pred = tree.predict(X_test)

mean_squared_error(y_test, pred)
```

Results:

Default: 6919.20814479638

Max_depth = 1 : 4903.77619310491

Min_Samples_Leaf = 10: 4443.06
