```python
def identify_constant_features(dataframe):
    count_uniques = dataframe.apply(lambda x: len(x.unique()))
    constants = count_uniques[count_uniques == 1].index.tolist()
    return constants
```

```python
constant_features_train = set(identify_constant_features(train))
print('There were {} constant features in TRAIN dataset.'.format(
        len(constant_features_train)))
```

```python
train.drop(constant_features_train, inplace=True, axis=1)
test.drop(constant_features_train, inplace=True, axis=1)
print('Train: {}\nTest: {}'.format(train.shape, test.shape))
```

```python
def identify_equal_features(dataframe):
    features_to_compare = list(combinations(dataframe.columns.tolist(),2))
    equal_features = []
    for compare in features_to_compare:
        is_equal = array_equal(dataframe[compare[0]],dataframe[compare[1]])
        if is_equal:
            equal_features.append(list(compare))
    return equal_features
```

```python
equal_features_train = identify_equal_features(train)
print('There were {} pairs of equal features in TRAIN dataset.'.format(len(equal_features_train)))
```

```python
features_to_drop = array(equal_features_train)[:,1] 
train.drop(features_to_drop, axis=1, inplace=True)
test.drop(features_to_drop, axis=1, inplace=True)
print('Train: {}\nTest: {}'.format(train.shape, test.shape))
```

```python
def identify_corr_features(dataframe):
    features_to_compare = list(combinations(dataframe.columns.tolist(),2))
    corr_features = []
    for compare in features_to_compare:
        corr = np.corrcoef(dataframe[compare[0]],dataframe[compare[1]])[0,1]
        if corr > 0.999 :
            corr_features.append(list(compare))
    return corr_features
```

```python
highCorr_features_train = identify_corr_features(train)
print('There were {} pairs of equal correlation in TRAIN dataset.'.format(len(highCorr_features_train)))
print('Pairs of equal correlation in TRAIN dataset{}.'.format(highCorr_features_train))
```

```python
X_im = train.drop(columns='TARGET')
```

```python
y_im = train.TARGET
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, n_estimators = 400, random_state=0)
clf.fit(X_im, y_im)

importances = clf.feature_importances_
indices = np.argsort(importances)[::-1]
columns = []

for i,v in enumerate(importances):
    print("%d. Feature %s(%.3f)" % (i + 1, X_im.columns.values[indices[i]], importances[indices[i]]))
    columns.append(X_im.columns.values[indices[i]])
```

```python
columns = columns[:50]
```

```python
train = train[columns]
```

```python
test = test[columns]
```

```python
train.shape, test.shape
```

```python
target = y_im

X_train = train[:30000]
X_valid = train[30000:]
y_train = target[:30000]
y_valid = target[30000:]
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import RandomizedSearchCV

gbm = LGBMClassifier(class_weight = 'balanced')

parameters_lgbm = {'learning_rate':(0.1,0.01), 'n_estimators':range(40,600,10),
                'max_depth':range(2,12,2),
              }

clf_lgbm = RandomizedSearchCV(gbm, parameters_lgbm, random_state=0, cv=5)

clf_lgbm.fit(X_train, y_train)
```

```python
y_pred_lgbm = clf_lgbm.predict_proba(X_valid)[:,1]
```

```python
average_precision_score(y_valid, y_pred_lgbm), roc_auc_score(y_valid, y_pred_lgbm)
```
