```python
base_numerical = base[['age','Medu', 'Fedu','traveltime', 'studytime',
       'failures', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences']]
```

```python
base_categorical = base[['school', 'address', 'famsize', 'Pstatus',
       'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'dicipline']]
```

```python
base_categorical.info()
```

```python
base_categorical.head()
```

```python
from sklearn.preprocessing import LabelEncoder

base_categorical = base_categorical.apply(LabelEncoder().fit_transform)
```

```python
base_categorical.head()
```

```python
base_all = pd.concat([base_categorical, base_numerical], axis = 1)
```

```python
base_all.head()
```

```python
X = base_all
y = base['G3']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, n_estimators = 100, random_state=0)
clf.fit(X, y)

importances = clf.feature_importances_
indices= np.argsort(importances)[::-1]

for i,v in enumerate(importances):
    print("%d. Feature %s(%.3f)" % (i + 1, X.columns.values[indices[i]], importances[indices[i]]))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.60, random_state=42)
```

```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import RandomizedSearchCV

lgbm = LGBMRegressor(n_jobs = -1)

parameters_lgbm = {'learning_rate':(0.1,0.01,0.001), 'n_estimators':range(40,400,10),
                'max_depth':range(-1,12,2)
              }

clf_lgbm = RandomizedSearchCV(lgbm, parameters_lgbm, cv=4)

clf_lgbm.fit(X_train, y_train)
prediction_lgbm = clf_lgbm.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, prediction_lgbm)
```

```python
from xgboost import XGBRegressor

xbg = XGBRegressor(n_jobs = -1)

parameters_xgb = {'learning_rate':(0.1,0.01,0.001), 'n_estimators':range(40,600,10),
                'max_depth':range(2,12,2)
              }

clf_xgb = RandomizedSearchCV(xbg, parameters_xgb, cv=4)

clf_xgb.fit(X_train, y_train)
prediction_xgb = clf_xgb.predict(X_test)
```

```python
mean_absolute_error(y_test, prediction_xgb)
```

```python
prediction = (prediction_xgb*0.5)+(prediction_lgbm*0.5)
mean_absolute_error(y_test, prediction)
```
